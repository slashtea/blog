## What is SpaCy and why you should use it!

* Hello, in this tutorial we're going to see what exactly is SpaCy and then do some NED(Nameed Entity Recognition) to Trump's speech in the UN General Assembly.

* It says **“We reject the ideology of globalism and accept the doctrine of patriotism.”** Let's find out !
* Full transcript can be found here: https://www.vox.com/2018/9/25/17901082/trump-un-2018-speech-full-text

```python
import spacy
from spacy import displacy
```

```python
with open('datasets/trump_us_speech.txt', 'r') as f:
    data = f.read()
```

```python
# load spacy.
nlp = spacy.load('en')
doc = nlp(data)
displacy.render(doc, style='ent', jupyter=True)
```

* Now what SpaCy did here is that it named entities with respect to their categories based on a pre trained model, this really useful to determine in a long document or a set of documents what the writer or speaker's opinion, topics addressed are. 

* Next we can aggregate those categories to see topics that Trump addressed the most in his speech.

#### Before we move on to doing that, there's another cool feature you can use which determines relationships between words in a sentence.

```python
from collections import Counter

cnt = Counter()
for ent in doc.ents:
    cnt[ent.label_] += 1

cnt.most_common(len(cnt))
```

* The list above are the most common named entities that Trump used is his speech. We can see that he used **GPE (Geopolitical entity)** very often, then **dates** and **ORG(organizations)** and last but not least **NORP (Nationalities or religious or political groups)**

* I guess that's standard political speech.
